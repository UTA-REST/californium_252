```python
# this make a Periodic Table plot you will neen boken
# I dont fully understand all parts so it will not be comented 
# simple changes should be managable from looking at it

from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.sampledata.periodic_table import elements
from bokeh.transform import dodge, factor_cmap
import pandas as pd
```

```python
# these were take from the Periodic_Table_Ionization notebook 

Xenon = ['Ca','Sr','Ba','La','Ta','W','Re','Os','Ir','Po','At','Rn','Ra','Ac','Ce',
         'Pr','Nd','Pm','Sm','Eu','Gd','Tb','Dy','Ho','Er','Tm','Th','Pa','U','Np',
         'Pu','Am','Cm','Bk','Cf','Es','Fm','Md','No']

Argon = ['Mg','Ca','Sc','Ti','V','Mn','Sr','Y','Zr','Nb','Tc','Sn','Ba','La','Hf',
         'Ta','W','Re','Os','Ir','Pb','Po','At','Rn','Ra','Ac','Ce','Pr','Nd','Pm',
         'Sm','Eu','Gd','Tb','Dy','Ho','Er','Tm','Yb','Lu','Th','Pa','U','Np','Pu',
         'Am','Cm','Bk','Cf','Es','Fm','Md','No']

Helium = ['Be','C','Mg','Al','Si','P','S','Cl','Ca','Sc','Ti','V','Cr','Mn','Fe','Co',
        'Ni','Cu','Zn','Ga','Ge','As','Se','Br','Kr','Sr','Y','Zr','Nb','Mo','Tc',
        'Ru','Rh','Pd','Ag','Cd','In','Sn','Sb','Te','I','Xe','Ba','La','Hf','Ta',
        'W','Re','Os','Ir','Pt','Au','Hg','Tl','Pb','Bi','Po','At','Rn','Ra','Ac',
        'Ce','Pr','Nd','Pm','Sm','Eu','Gd','Tb','Dy','Ho','Er','Tm','Yb','Lu','Th',
        'Pa','U','Np','Pu','Am','Cm','Bk','Cf','Es','Fm','Md','No']

Californium = ['Sb','Sr','Kr','Ce','Br','Te','Eu','Rb','Mo','Y','Rh','Pr','Xe','Pm',
               'La','Ba','Pd','Tc','Ru','Nb','Cd','I','Nd','Sm','Zr','Cs','Sn','Ag']
```

```python
output_file("periodic.html")

periods = ["I", "II", "III", "IV", "V", "VI", "VII","LA","AC"]
groups = [str(x) for x in range(1, 19)]

df = elements.copy()
df["atomic mass"] = df["atomic mass"].astype(str)
df["group"] = df["group"].astype(str)
df["period"] = [periods[x-1] for x in df.period]

df.iloc[70, df.columns.get_loc('group')] = str(3)
df.iloc[70, df.columns.get_loc('period')] = "VI"
df.iloc[102, df.columns.get_loc('group')] = str(3)
df.iloc[102, df.columns.get_loc('period')] = "VII"
for q in range(56,70):
    ll = q - 56 +3
    df.iloc[q, df.columns.get_loc('group')] = str(ll)
    df.iloc[q, df.columns.get_loc('period')] = "LA"
for q in range(88,102):
    ll = q - 88 +3
    df.iloc[q, df.columns.get_loc('group')] = str(ll)
    df.iloc[q, df.columns.get_loc('period')] = "AC"

for q in range(0,len(df)):
    elem = df.iloc[q]["symbol"]
    if (elem in Helium):
        df.iloc[q, df.columns.get_loc('metal')] = "MEMEME"
        
cmap = {
    "alkali metal"         : "#DFDFDF",
    "alkaline earth metal" : "#DFDFDF",
    "metal"                : "#DFDFDF",
    "halogen"              : "#DFDFDF",
    "metalloid"            : "#DFDFDF",
    "noble gas"            : "#DFDFDF",
    "nonmetal"             : "#DFDFDF",
    "transition metal"     : "#DFDFDF",
    "lanthanoid"           : "#DFDFDF",
    "actinoid"             : "#DFDFDF",
    "MEMEME"               : "#599d7A"
}


TOOLTIPS = [
    ("Name", "@name"),
    ("Atomic number", "@{atomic number}"),
    ("Atomic mass", "@{atomic mass}"),
    ("Type", "@metal"),
    ("CPK color", "$color[hex, swatch]:CPK"),
    ("Electronic configuration", "@{electronic configuration}"),
]


p = figure( plot_width=1000, plot_height=450,
           x_range=groups, y_range=list(reversed(periods)),
           tools="hover", toolbar_location=None, tooltips=TOOLTIPS)

r = p.rect("group", "period", 0.95, 0.95, source=df, fill_alpha=0.6, 
           color=factor_cmap('metal', palette=list(cmap.values()), factors=list(cmap.keys())))

text_props = {"source": df, "text_align": "left", "text_baseline": "middle"}

x = dodge("group", -0.4, range=p.x_range)

p.text(x=x, y="period", text="symbol", text_font_style="bold", **text_props)

p.text(x=x, y=dodge("period", 0.3, range=p.y_range), text="atomic number",
       text_font_size="8pt", **text_props)

p.text(x=x, y=dodge("period", -0.35, range=p.y_range), text="name",
       text_font_size="5pt", **text_props)

p.text(x=x, y=dodge("period", -0.2, range=p.y_range), text="atomic mass",
       text_font_size="5pt", **text_props)



df = elements.copy()
df["atomic mass"] = df["atomic mass"].astype(str)
df["group"] = df["group"].astype(str)
df["period"] = [periods[x-1] for x in df.period]

df.iloc[70, df.columns.get_loc('group')] = str(3)
df.iloc[70, df.columns.get_loc('period')] = "VI"
df.iloc[102, df.columns.get_loc('group')] = str(3)
df.iloc[102, df.columns.get_loc('period')] = "VII"
for q in range(56,70):
    ll = q - 56 +3
    df.iloc[q, df.columns.get_loc('group')] = str(ll)
    df.iloc[q, df.columns.get_loc('period')] = "LA"
for q in range(88,102):
    ll = q - 88 +3
    df.iloc[q, df.columns.get_loc('group')] = str(ll)
    df.iloc[q, df.columns.get_loc('period')] = "AC"
#df.loc[(df["period"] == 'VI'),'period']='VV'

for q in range(0,len(df)):
    elem = df.iloc[q]["symbol"]
    if (elem in Californium):
        df.iloc[q, df.columns.get_loc('metal')] = "MEMEME"
df2 = df[df.metal == 'MEMEME']
text_props = {"source": df2, "text_align": "left", "text_baseline": "middle"}
p.text(x=x, y="period", text="symbol", text_font_style="bold",text_color="#BB0000", **text_props)


p.outline_line_color = None
p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_standoff = 0

show(p)
```
