```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data_file = np.loadtxt("Cf_data/Cf_raw.txt",skiprows=4)
Z_number  = Data_file[:,0]
N_number  = Data_file[:,1]
Yeild     = Data_file[:,2]
```

```python
plt.figure(figsize=(8,7))

bins = (np.unique(N_number),np.unique(Z_number))
plt.hist2d(N_number, Z_number, weights=Yeild*100,bins=bins,cmap='Reds')

plt.xlabel("N",fontsize=20)
plt.ylabel("Z",fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.ylim(28,68)
plt.xlim(38,108)

plt.grid()
plt.clim(0.1,5)
cb = plt.colorbar()#.set_label(label='Yeild',size=20).tick_params(labelsize=10) 
cb.set_label(label='Yeild',size=20)
cb.ax.tick_params(labelsize=15) 
#plt.tight_layout()
#plt.savefig("Plots/Cf_Fission.png",dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,7))
plt.scatter(N_number,Z_number,c = Yeild*100,s=75,marker='s')
plt.xlabel("N",fontsize=15)
plt.ylabel("Z",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.ylim(28,68)
plt.xlim(38,108)

plt.grid()
plt.colorbar()

plt.show()
```
